```python
data = pd.read_csv('Twitter-sentiment-self-drive-DFE.csv', encoding=("ISO-8859-1"), low_memory=False)
```

```python
fig = px.scatter(data, x='sentiment',y='sentiment:confidence')
fig.show()
```

```python
response = len(data.text[0])
response
```

```python
rlist = data.text.tolist()
```

```python
def getlen():
    
    lengthlist=[]
    
    for x in rlist:
        lengthlist.append(len(x))
        
    return lengthlist
```

```python
getlen()
```

```python
data['textlength']=getlen()
```

```python
data.head()
```

```python
fig = px.scatter(data, x='sentiment:confidence',y='textlength')
fig.show()
```
